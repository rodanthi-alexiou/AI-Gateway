### 🗑️ Clean up resources

When you're finished with the lab, you should remove all your deployed resources from Azure to avoid extra charges and keep your Azure subscription uncluttered.

The following command runs `terraform destroy` to remove every resource created by this lab.

```python
import os, sys
sys.path.insert(1, '../../shared')  # add the shared directory to the Python path
import utils

output = utils.run("az account show", "Retrieved az account", "Failed to get the current az account")

subscription_id = ''
if output.success and output.json_data:
    subscription_id = output.json_data['id']

# Specify the target subscription for Terraform
os.environ['ARM_SUBSCRIPTION_ID'] = subscription_id

output = utils.run(
    "terraform destroy -auto-approve",
    "Resources deleted successfully",
    "Resources deletion failed"
)
```
